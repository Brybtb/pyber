## Total Fares by City Type

```python
# Calculate Type Percents
# Calculate Total Fare by City Type
fare = df.groupby(['type'])['fare'].sum() 

# Labels for the sections of our pie chart
labels = ['Rural', 'Suburban', 'Urban']

# The colors of each section of the pie chart
colors = ['orange','blue','red']
explode = (0, 0, 0.1)

# Build Pie Chart
plt.title('Percentage of Total Fares by City Type')
plt.pie(fare, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%',shadow=True, startangle=160)
plt.axis('equal')
#plt.savefig('Images/% of Total Fares By City Types.png')
plt.show()


# Save Figure
plt.savefig('Percentage of Total Fares by City Type')
```

```python
# Show Figure
plt.show()
```

## Total Rides by City Type

```python
# Calculate Total Fare by City Type
totz = df.groupby(['type'])['ride_id'].count()

# Labels for the sections of our pie chart
labels = ['Rural', 'Suburban', 'Urban' ]

# The colors of each section of the pie chart
colors = ['orange','blue','red']
explode = (0, 0, 0.1)
plt.title('Total Rides by City Type')
plt.pie(totz, explode=explode, labels=labels, colors=colors,
        autopct='%1.1f%%', shadow=True, startangle=140)
plt.axis('equal')
plt.savefig('Total Rides by City Type')
plt.show()
```

```python
# Show Figure
plt.show()
```

## Total Drivers by City Type

```python
# Calculate Driver Percents

total_rides = df.groupby(['type'])['ride_id'].count()

# Labels for the sections of our pie chart
labels = ['Rural','Suburban','Urban' ]

# The colors of each section of the pie chart
colors = ['orange','blue','red']
explode = (0, 0, 0.1)
plt.title('% of Total Rides By City Types')
plt.pie(total_rides, explode=explode, labels=labels, colors=colors,
        autopct='%1.1f%%', shadow=True, startangle=140)
plt.axis('equal')
plt.savefig('Total Drivers by City Type')
plt.show()
# Build Pie Charts

# Save Figure
plt.savefig('Total Drivers by City Type')
```
